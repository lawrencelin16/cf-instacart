# 02. Creating the price range and busiest day columns

```python
# Create a subset
df = ords_prods_merged[:1000000]
df.shape
```

```python
# Define the function using if statements 
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# Step 4
# Create a new column to classify the busyness of periods during the day

# Examine the frequency of orders by hour of day.
ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

Based on this information, orders per hour range from 0 to 2,761,760. The following criteria will be created:

Fewest orders: 0 to 500,000 orders

Average orders: 500,001 to 2,500,000 orders

Most orders: 2,500,001+ orders


That leaves us with the following grouping:

Fewest orders: 3, 4, 2, 5, 1, 0, 6, 23

Most orders: 10, 11, 14, 15, 13, 12, 16, 9

Average orders: The remaining hours (17, 8, 18, 19, 20, 7, 21, 22)

```python
# Use a for-loop to create the new column
result = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value in [3, 4, 2, 5, 1, 0, 6, 23]:
    result.append("Fewest orders")
  elif value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result.append("Most orders")
  else:
    result.append("Average orders")

ords_prods_merged['busiest_period_of_day'] = result
```

```python
# Step 5
# Print the frequency of the new column
ords_prods_merged['busiest_period_of_day'].value_counts()
```
